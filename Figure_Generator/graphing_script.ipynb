```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# here we import the data from the json file
with open("Network_Statistics") as f:
    network_statistics = json.load(f)
```

```python
# average resistance plot with shaded portions for standard deviation of the average

resistances_300_std = np.array(network_statistics["300 nodes"]["resistances"]["std"])
resistances_200_std = np.array(network_statistics["200 nodes"]["resistances"]["std"])
resistances_100_std = np.array(network_statistics["100 nodes"]["resistances"]["std"])

resistances_300_mean = np.array(network_statistics["300 nodes"]["resistances"]["mean"])
resistances_200_mean = np.array(network_statistics["200 nodes"]["resistances"]["mean"])
resistances_100_mean = np.array(network_statistics["100 nodes"]["resistances"]["mean"])

plt.plot(resistances_300_mean, label="300 node networks")
plt.fill_between(
    range(len(resistances_300_mean)),
    (resistances_300_mean - resistances_300_std),
    (resistances_300_mean + resistances_300_std),
    color="b",
    alpha=0.1,
)
plt.plot(resistances_200_mean, label="200 node networks")
plt.fill_between(
    range(len(resistances_200_mean)),
    (resistances_200_mean - resistances_200_std),
    (resistances_200_mean + resistances_200_std),
    color="orange",
    alpha=0.1,
)
plt.plot(resistances_100_mean, label="100 node networks")
plt.fill_between(
    range(len(resistances_100_mean)),
    (resistances_100_mean - resistances_100_std),
    (resistances_100_mean + resistances_100_std),
    color="g",
    alpha=0.1,
)
plt.xlabel("Lloyd Iteration")
plt.ylabel("Resistance")
plt.title("Mean Resistance Per Lloyds Iteration")
plt.legend(bbox_to_anchor=(1.40, 0.9), loc="center right")
plt.show()
```
